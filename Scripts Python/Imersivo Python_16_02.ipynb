```python
df = pd.merge(df_vendas, df_produtos, how= "left", on= "ID Produto")
```

```python
df["Custo"] = df["Custo Unitário"]*df["Quantidade"]
```

```python
df['Lucro'] = df['Valor Venda'] - df["Custo"]
```
